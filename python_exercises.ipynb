```python
import pandas as pd
import numpy as np

# Create sample dataset
np.random.seed(42)
n_students = 200

data = {
    'student_id': range(1000, 1000 + n_students),
    'major': np.random.choice(['CS', 'Math', 'Physics', 'Biology'], n_students),
    'year': np.random.choice([1, 2, 3, 4], n_students),
    'exam_score': np.random.normal(75, 10, n_students).clip(0, 100),
    'assignments_completed': np.random.randint(0, 11, n_students),
    'hours_studied': np.random.normal(15, 5, n_students).clip(1, 40)
}

df = pd.DataFrame(data)

# Introduce some NaN values
df.loc[np.random.choice(n_students, 10), 'exam_score'] = np.nan
df.loc[np.random.choice(n_students, 5), 'hours_studied'] = np.nan

#Provides brief description of data
df.describe()

df.isnull().sum()

df['exam_score'] = df['exam_score'].fillna(df.groupby('major')['exam_score'].transform('mean'))

df['hours_studied'] = df['hours_studied'].fillna(df.groupby('year')['hours_studied'].transform('median'))


average_major_score = df.groupby('major')['exam_score'].mean()

print("Averave Score by Major:")
print(average_major_score)


# Finding major with highest score
highest_major = average_major_score.idxmax()

print("Major with highest score:")
print(highest_major)

# Finding correlation between hours_studied and exam_score
correlation = df['hours_studied'].corr(df['exam_score'])

print("Correlation between hours studied and student exam score:")
print(correlation)

#Defining bins and labels for performance column
bins = [0, 69.99, 80, 90, 100]
labels = ['Needs Improvement', 'Average', 'Good', 'Excellent']
df['performance'] = pd.cut(df['exam_score'], bins=bins, labels=labels)


#Task 3

#Calculating average hours studied, average exam score, number of students for each major and year
combined_stats = df.groupby(['major', 'year']).agg(
    student_num = ('student_id', 'count'),
    average_score =('exam_score', 'mean'),
    average_hours=('hours_studied', 'mean')

)

#Top 5 Students with highest scores

top_5_students = df.nlargest(5, 'exam_score', keep='all')

#Creating pivot table

pivot = pd.pivot_table(
    df,
    values = 'exam_score',
    index='major',
    columns = 'year',
    aggfunc='mean')



```
